```python
import pandas as pd

# Load key datasets
teams = pd.read_csv("../data/MTeams.csv")  # Men's teams
tourney_results = pd.read_csv("../data/MNCAATourneyCompactResults.csv")  # NCAA Tournament Results
regular_season = pd.read_csv("../data/MRegularSeasonCompactResults.csv")  # Regular season results
seeds = pd.read_csv("../data/MNCAATourneySeeds.csv")  # Tournament seeds

# Quick inspection
print("Teams Data:")
print(teams.head(), "\n")

print("Tournament Results:")
print(tourney_results.head(), "\n")

print("Regular Season Results:")
print(regular_season.head(), "\n")

print("Tournament Seeds:")
print(seeds.head(), "\n")
```

Explore the Data. Check for Missing Data

```python
print("Missing values in Teams dataset:")
print(teams.isnull().sum(), "\n")

print("Missing values in Tournament Results dataset:")
print(tourney_results.isnull().sum(), "\n")

print("Missing values in Regular Season dataset:")
print(regular_season.isnull().sum(), "\n")

print("Missing values in Tournament Seeds dataset:")
print(seeds.isnull().sum(), "\n")
```

 Explore Summary Statistics
 This gives us insights into the distribution of scores, win margins, and trends.

```python
print("Tournament Results Summary:")
print(tourney_results.describe(), "\n")

print("Regular Season Results Summary:")
print(regular_season.describe(), "\n")
```

# Visualize Key Stats

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.histplot(tourney_results["WScore"], bins=30, kde=True)
plt.title("Distribution of Winning Scores (NCAA Tournament)")
plt.xlabel("Winning Score")
plt.ylabel("Frequency")
plt.show()
```

```python
tourney_results["WinMargin"] = tourney_results["WScore"] - tourney_results["LScore"]

plt.figure(figsize=(10, 5))
sns.histplot(tourney_results["WinMargin"], bins=30, kde=True)
plt.title("Distribution of Win Margins")
plt.xlabel("Win Margin")
plt.ylabel("Frequency")
plt.show()
```

```python
regular_season["WLoc"].value_counts().plot(kind="bar", color=["blue", "green", "red"])
plt.title("Winning Locations (H=Home, A=Away, N=Neutral)")
plt.xlabel("Location")
plt.ylabel("Game Count")
plt.show()
```
